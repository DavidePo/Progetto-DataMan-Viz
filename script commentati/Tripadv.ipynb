```python
#importazione delle librerie necessarie per calcolare la distanza tra le varie località e ottenere musei che siano nel raggio di 25 km
import numpy
import scipy
import reverse_geocode
import geopy.distance
```

```python
lista_citta = {"Amsterdam" : [52.22 , 4.53], "Athens" : [37.58 , 23.42] , "Barcelona" : [41.22 , 2.10] , "Berlin" : [52.30  , 13.25] ,
       "Bratislava" : [48.09 , 17.07],  "Brussels"  : [50.50 , 4.20] , "Budapest" : [47.3 , 19.05] , "Copenhagen" : [55.4 , 12.35]  ,
       "Dublin"  : [53.20 , 6.15]  , "Edinburgh" : [55.57 , -3.11]   ,  "Geneva"   : [46.12 , 6.09]  ,  "Hamburg" : [53.34 , 10.0] , "Helsinki" : [60.10 , 24.58] ,             
       "Krakow" : [50.3 , 19.56] ,  "Lisbon" : [38.43 , 9.08] ,  "Ljubljana" : [46.03 , 14.31] ,  "London" : [51.3 , 0.10] , "Luxembourg" : [49.45 , 6.05],
       "Lyon" : [45.46 , 4.50] , "Madrid" : [40.24, 3.41], "Milan" : [45.28, 9.10]  ,  "Munich" : [48.08 , 11.34],   
        "Oporto"  : [41.09 , 8.38], "Oslo" : [59.55 , 10.45] ,  "Paris" : [48.52 , 2.20] ,
       "Prague" : [50.05 , 14.26], "Rome" : [41.54 , 12.29],"Stockholm" : [59.20 , 18.03],"Vienna" : [48.13 , 16.20],"Warsaw" : [52.13 , 21.00],"Zurich" : [47.22 , 8.32]}
```

```python
lista_citta_finale = []
lista_index = []
```

```python
for chiave in lista_citta:
    for index, row in data_ridotto_per_analisi.iterrows():
        coordinates = (row['Latitude'],row['Langtitude'])
        coordinates2 = (lista_citta[chiave][0],lista_citta[chiave][1])
        if geopy.distance.distance(coordinates, coordinates2).km < 25:
            lista_citta_finale.append(chiave)
            lista_index.append(index)
            
```

```python
data_museum = data_ridotto_per_analisi.loc[lista_index]
```

```python
data_museum['city'] = lista_citta_finale
```

```python
data_museum
```

```python
#creazione di variabili necessarie per la media pesata e rating successivo
data_museum['molt'] = data_museum['Rating']*data_museum['ReviewCount']
```

```python
sum_review = data_museum.groupby('city').sum()['ReviewCount']
```

```python
sum_tot = data_museum.groupby('city').sum()['molt']
```

```python
sum_tot/sum_review
```

```python
data_gb = pd.DataFrame()
```

```python
data_gb['sum_review'] = sum_review 
data_gb['molt'] = sum_tot
```

```python
data_gb['media_pesata'] = data_gb['molt']/data_gb['sum_review']
data_gb
```

```python
#dataset finale 
data_museum_def = data_gb['media_pesata']
data_museum_def = pd.DataFrame(data_museum_def).reset_index()
data_museum_def
```

```python
#caricamento su mongo
museum_rating = db.museum_rating
museum_rating.insert_many(data_museum_def.to_dict('records'))
```

```python
#stesso procedimento per ristoranti
rest = pd.read_csv('TA_restaurants_curated.csv')
rest = rest.drop_duplicates()
```

```python
rest_ridotto = rest[['City','Rating','Number of Reviews']]
rest_ridotto = rest_ridotto.dropna()
```

```python
tot_reviews = rest_ridotto.groupby('City').sum()['Number of Reviews']
```

```python
rest_ridotto['media_pesata'] = rest_ridotto['Rating']*rest_ridotto['Number of Reviews']
```

```python
pesi = rest_ridotto.groupby('City')['media_pesata'].sum()
```

```python
data_finale = pd.DataFrame(tot_reviews)
```

```python
data_finale['pesi'] = pesi
```

```python
data_finale['media_pesata'] = data_finale['pesi']/data_finale['Number of Reviews']
```

```python
restaurant = data_finale.reset_index()
restaurant.columns = ['city', 'Number of Reviews', 'pesi', 'media_pesata']
restaurant
```

```python
#il dataframe finale contiene 31 città, caricamento su mongo
restaurant_rating = db.restaurant_rating
restaurant_rating.insert_many(restaurant.to_dict('record'))
```
