```python
#importare librerie necessarie per scraping
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import requests
```

```python
#creazione di una lista con le città di interesse
city_list = ['Amsterdam', 'Atene', 'Barcellona', 'Berlino', 'Bratislava',
             'Bruxelles', 'Budapest', 'Copenhagen', 'Dublino', 'Edinburgh',
             'Ginevra', 'Amburgo', 'Helsinki', 'Lisbona','Ljubljana','Londra',
             'Lussemburgo', 'Lione', 'Madrid', 'Milano', 'Monaco', 'Porto', 'Oslo',
             'Parigi', 'Praga', 'Roma', 'Stoccolma', 'Vienna', 'Varsavia', 'Zurich']
```

```python
#creazione di un ciclo che appende ad una lista i dati per il costo della vita delle città di interesse
base_url = 'https://numbeo.com/cost-of-living/in/'
list_data_complete = []


for city in city_list:
    print(city)
    url = base_url + city + '?displayCurrency=EUR'
    response = requests.get(url)
    html_page = response.text
    html_tree = BeautifulSoup(html_page)
    
    lista = []

    for row in html_tree.find_all('tr')[1:]:
        temp = row.get_text().replace('\n', '')
        temp = temp.replace('\xa0', ' ')
        lista.append(temp)

    lista = lista[2:-4]

    
    l = []
    
    for el in lista:
        l.append(el.split('  '))
    for el in l:
        if len(el) > 2:
            del el[1]

    data = pd.DataFrame(l, columns = ['index','value'])
    
    indicatori_base = data[0:6]
    indicatori_costo_vita = data[6::].reset_index()
    data1 = pd.DataFrame(indicatori_base['index'].str.split(':').tolist())
    data1.columns = ['index','value']
    
    lista_value = pd.DataFrame(indicatori_costo_vita['value'].str.split('€').tolist())[0]
    
    data2 = pd.DataFrame(indicatori_costo_vita['index'])
    data2['value'] = lista_value
    
    data2 = data2[~data2['value'].str.contains("Edit")]
    
    data_complete = data1.append(data2)
    list_data_complete.append(data_complete)
    
    
    
```

```python
list_data_complete[0]['valore'] #stampa valori Amsterdam
```

```python
#creazione dataframe
cost_of_life_df = list_data_complete[0]

i = 1
l = 0
for i in list_data_complete:
    cost_of_life_df[city_list[l]] = i['valore']
    l += 1
    
```
