```python
#Creo una lista contente tutti i DataFrame che riguardano gli indici di Eurostat
list_df = []

for collection in  eu_db.list_collection_names():
    coll = eval('eu_db.' + collection)
    cursor = coll.find({})
    list_df.append(pd.DataFrame(list(cursor)))
```

```python
#Elimino tutte le colonne '_id' ed i duplicati dai DataFrame di Eurostat
list_no_dp = []

for df in list_df:
    del df['_id']
    df = df.drop_duplicates()
    list_no_dp.append(df)
```

```python
#Creo un Dataset unico in cui unisco tutti i Dataset di Eurostat usando come chiave la colonna 'city' 
from functools import reduce
data_eurostat_integrated = reduce(lambda x, y: pd.merge(x, y, on = 'city', how='outer'), list_no_dp)
data_eurostat_integrated
```
